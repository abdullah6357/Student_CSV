### 2.Data Cleaning

### 3.Split the Data into Two Data Frames One would contain Student Infromation and The Other would contain only scores ,id would be common)

```python
std_info = df.loc[:, :'career_aspiration']
std_score = df.loc[:, 'math_score':'physics_score']
```

```python
std_score.head(7)
```

### 4.To Park unknown Records in Separate Data Frames, These should be removed from Student Information and Student Score Data Frames

```python
sp = ['Unknown']
std_unkno = df[df['career_aspiration'].isin(sp)]
std_unkno.head(7)
# std-unkno = df.where(df['career_aspiration'] == 'Unknown') another method.
```

```python
std_info.drop(std_info[std_info['career_aspiration'] == 'Unknown'].index, inplace=True)
std_info.head(7)
```

### 5.Join These Two Data Frames  Student Inforamation and Student Score into single data frame (Student Consolidated)

```python
# df1.join(df2 , on = 'FellowshipID' , how = 'outer' , lsuffix = '_Left' , rsuffix = '_Right' )
```

```python
std_consolidated = std_info.join(std_score , on = 'id' , how = 'inner',lsuffix = '_Left' , rsuffix = '_Right')
std_consolidated.head(7)
```

### 6.Minimum, Maximum and Average Math Scores by Career Aspiration, I.e Lawer etc

```python
std_consolidated.groupby('career_aspiration').agg({'math_score': ['min','max','mean'] })
```

### 7.Average score of Student with Name Statting with ‘P’

```python
std_p = std_consolidated[std_consolidated['first_name'].str.contains('P')]
std_p .head(7)
```

```python
std_p['avg_score'] = std_p[['math_score', 'history_score', 'physics_score']].mean(axis = 1)
std_p.head(7)
```

### 8.Average Chemistery Score of Students that have achieved more than 50 Marks in Maths

### 9.Return the Student Name with 50th Highest Maths Score

```python
marks = std_consolidated[['first_name', 'math_score']]
marks = marks.copy()

marks['rank'] = marks[['math_score']].rank(method='dense', ascending=False)

marks = marks[marks['rank'] == 1].head(50)

marks = marks.sort_values('id', ascending=False).head(1)

marks
```

### 13.Apply a case Statement of Student Total Score and categorize into Low, Mid, High

```python
case = std_consolidated[['first_name', 'career_aspiration', 'Total Score']]
case = case.copy()

for idx, row in case.iterrows():
    if row['Total Score'] in range(230,301):
        case.loc[idx, 'category'] = 'high'
    elif row['Total Score'] in range(75,229):
        case.loc[idx, 'category'] = 'mid'
    else:
        case.loc[idx, 'category'] = 'low'

case.head(7)
```
